```python
import pandas as pd
import json

# Load the JSON file
with open('/content/DataEngineeringQ2.json') as file:
    data = json.load(file)

# Normalize the JSON data
df = pd.json_normalize(data, sep='_')
```

```python
df
```

```python
patient_details_list = [item['patientDetails'] for item in data]
```

```python
# Convert the list to a DataFrame
patient_details = pd.DataFrame(patient_details_list)
```

```python
patient_details
```

```python
def missing_percentage(column):
    total = len(df)
    missing = df[column].apply(lambda x: x == '').sum()
    return round((missing / total) * 100, 2)
```

```python

def missing_percentage(column):
    total = len(patient_details)
    missing = patient_details[column].apply(lambda x: x == '' or pd.isnull(x)).sum()
    return round((missing / total) * 100, 2)


missing_percentages = {
    'firstName': missing_percentage('firstName'),
    'lastName': missing_percentage('lastName'),
    'DOB': missing_percentage('birthDate')
}

missing_percentages
```

```python

mode_gender = patient_details['gender'].mode()[0]

patient_details['gender'].replace('', mode_gender, inplace=True)
patient_details['gender'].fillna(mode_gender, inplace=True)
```

```python

total = len(patient_details)
female_count = (patient_details['gender'] == 'Female').sum()
female_percentage = round((female_count / total) * 100, 2)
```

```python
female_percentage
```

```python
import datetime

patient_details['birthDate'] = pd.to_datetime(patient_details['birthDate'], errors='coerce')

current_year = datetime.datetime.now().year
patient_details['age'] = patient_details['birthDate'].apply(lambda x: current_year - x.year if pd.notnull(x) else None)
```

```python

def categorize_age(age):
    if age is None:
        return 'Unknown'
    elif age <= 12:
        return 'Child'
    elif 13 <= age <= 19:
        return 'Teen'
    elif 20 <= age <= 59:
        return 'Adult'
    else:
        return 'Senior'

patient_details['ageGroup'] = patient_details['age'].apply(categorize_age)
```

```python

adult_count = patient_details[patient_details['ageGroup'] == 'Adult'].shape[0]
adult_count
```

```python
from collections import Counter
```

```python
medicines_frozensets = [frozenset(med.items()) for med in medicines]

# Count the frequency of each medicine
medicine_counter = Counter(medicines_frozensets)

# Get the 3rd most common medicine
most_common = medicine_counter.most_common(3)
print(most_common[2] if len(most_common) > 2 else "Not enough data")
```

```python

statuses = [med['status'] for med in df['consultationData_medicines']]

status_counter = Counter(statuses)

total_medicines = len(df['consultationData_medicines'])

active_percentage = (status_counter['active'] / total_medicines) * 100
inactive_percentage = (status_counter['inactive'] / total_medicines) * 100

print(f"{active_percentage:.2f}, {inactive_percentage:.2f}")
```

```python
def is_valid_indian_phone_number(phone_number):

    phone_number = phone_number.strip()

    if phone_number.startswith('+91'):
        phone_number = phone_number[3:]
    elif phone_number.startswith('91'):
        phone_number = phone_number[2:]

    if len(phone_number) == 10 and phone_number.isdigit():

        if phone_number.isdigit() and 6000000000 <= int(phone_number) <= 9999999999:
            return True
    return False

dataset = df

for record in dataset:
    record['isValidMobile'] = is_valid_indian_phone_number(record['phoneNumber'])

valid_count = sum(1 for record in dataset if record['isValidMobile'])

for record in dataset:
    print(record)

print(f"Number of valid phone numbers found: {valid_count}")
```
